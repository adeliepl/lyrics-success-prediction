<h3> Defining Success </h3>

```python
df = data
```

```python
numeric_columns = [
    "Week Position", "Weeks on Chart", "spotify_track_duration_ms", "danceability",
    "energy", "key", "loudness", "mode", "speechiness", "acousticness",
    "instrumentalness", "liveness", "valence", "tempo", "time_signature",
    "spotify_track_popularity"
]

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
df_grouped = df.groupby("SongID").agg({
    "Weeks on Chart": "mean",
    "spotify_track_popularity": "mean",
    "Week Position": "min",  # Best position achieved
}).reset_index()
```

```python
average_popularity = df_grouped["spotify_track_popularity"].mean()
average_weeks_on_chart = df_grouped["Weeks on Chart"].mean()
```

```python
top_songs = df_grouped[df_grouped["Week Position"] <= 10]
average_weeks_in_top_10 = top_songs["Weeks on Chart"].mean()
```

```python
hit_threshold_popularity = average_popularity * 1.2  # 20% above average
hit_threshold_weeks = average_weeks_in_top_10  # Average weeks in Top 10
```

```python
average_popularity
```

```python
average_weeks_on_chart
```

```python
average_weeks_in_top_10
```

```python
def label_hit_song(row):
    if (row["spotify_track_popularity"] >= hit_threshold_popularity) and (
        row["Weeks on Chart"] >= hit_threshold_weeks):
        return True
    return False

df["Hit"] = df.apply(label_hit_song, axis=1)
```
